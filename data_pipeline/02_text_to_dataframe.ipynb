```python
from pathlib import Path

import pandas as pd
```

```python
source_folder_path = Path(
    r"/Users/uuboy.scy/side-project/tibame/tjr103/data_pipeline/res_gossiping",
)
for txt_path in source_folder_path.glob("*.txt"):
    print(txt_path)
```

```python
data = []
for txt_path in source_folder_path.glob("*.txt"):
    with txt_path.open("r", encoding="utf-8") as f:
        data.append(f.read().split("---split---")[1].split("\n")[1:-1])
```

```python
columns = [
    "推",
    "噓",
    "分數",
    "作者",
    "標題",
    "時間",
]
```

```python
df = pd.DataFrame(
    data=data,
    columns=columns,
)

df
```

```python
def convert_score_str_to_int(input_str: str) -> int:
    """Convert `推: 25` -> 25"""
    return int(input_str.split(":")[1].strip())
```

```python
df["推_t"] = df["推"].apply(convert_score_str_to_int)
df
```
